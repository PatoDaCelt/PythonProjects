GOLES DE RONALDO, MESSI Y SUÁREZ EN 2015

.isin()  Se utiliza para verificar si los valores de una columna están dentro de una lista o conjunto de valores.

```python
players = ['Cristiano Ronaldo', 'Lionel Messi', 'Luis Suárez']
top_players = df3[(df3.player_name.isin(players)) & (df3.year == 2015)]
cols = ['player_name', 'year', 'goals']
top_players[cols]
```

.loc[filas, columnas]   Permite seleccionar datos por etiquetas (nombres, no posiciones).

.query()  Permite filtrar filas usando una expresión booleana en forma de string, como si escribieras una condición SQL o matemática.

```python
players = ['Cristiano Ronaldo', 'Lionel Messi', 'Luis Suárez']
cols = ['year', 'goals']
top_players = df3.loc[players, cols]
top_players.query('year==2015')
```

.sum() → suma todos los valores de una columna.

.mean() → promedio de los valores de una columna.

.max() → valor máximo de una columna.

.min() → valor mínimo de una columna.

.cumsum() → en cada fila va poniendo la suma acumulada de una columna.

.cummax() → en cada fila va poniendo el valor máximo encontrado en orden de una columna.

.value_counts(sort=True) → cuenta los distintos valores que existen en una columna (muy útil para contar categorías) y los ordena descendientemente.

.drop_duplicates(subset='col_name') → elimina duplicados en una columna.

```python
df3['goals'].sum()
df3['goals'].mean()
df3['goals'].cumsum()
df3['goals'].cummax()
df3['goals'].value_counts(sort=True)
df3.drop_duplicates(subset='goals')
```

MÁXIMA CANTIDAD DE GOLES HECHOS POR UN SOLO JUGADOR EN UNA SOLA TEMPORADA

.agg(function)   Permite pasar funciones que nosotros mismos hayamos creado u otras funciones importadas de otras librerías

```python
max_goals = df3['goals'].max()
cols = ['goals', 'year', 'team_name']
df3[df3.goals == max_goals][cols]
```

PORCENTAJE DE DISPAROS AL ARCO QUE TERMINAN EN GOL

```python
total_shots = df3['shots'].sum()  #Sumatoria de shots
total_goals = df3['goals'].sum()  #Sumatoria de goals
round(total_goals / total_shots * 100, 2)  #Redondeo del porcentaje a 2 decimales
```

TOP 10 JUGADORES QUE MÁS GOLES HAN MARCADO

.groupby  Agrupa un DataFrame por los valores de una o más columnas

    df3.groupby('player_name'): Agrupa las filas por el valor en la columna 'player_name'.

    ['goals']: Selecciona solo la columna 'goals' sobre la cual se hará el cálculo.

    .sum(): Suma los goles de cada jugador.

```python
top_players = df3.groupby('player_name')['goals'].sum()
top_players.sort_values(ascending=False).head(10)
```

TOP 5 EQUIPOS CON MÁS TARJETAS ROJAS Y AMARILLAS

```python
cards = df3.groupby('team_name')[['red_cards', 'yellow_cards']].sum()
cards.sort_values('red_cards', ascending=False).head(5)
```

Crear una columna nueva

TOP 5 MEJORES JUGADORES DE TODO EL TIEMPO

```python
df3['goals_assists'] = df3.goals + df3.assists  #Crea una columna nueva a partir de la suma de otras
best_players = df3.groupby('player_name')['goals_assists'].sum()
best_players.sort_values(ascending=False).head()
```

EJERCICIO #5
¿Qué porcentaje de goles son hechos por penales?
La columna npg significa: goles hechos SIN penales.

```python
total_penaltis = df3['goals'].sum() - df3['npg'].sum()
total_goals = df3['goals'].sum()
penaltis_porcent = round((total_penaltis / total_goals) * 100, 2)
penaltis_porcent
```

EJERCICIO #6
¿Cuál fue el equipo que más goles hizo cada año? 

```python
goals_per_year = df3.groupby(['year', 'team_name'])['goals'].sum().reset_index()
most_goals_per_year_team =  goals_per_year.sort_values(['year','goals'], ascending=[True,False]).groupby('year').first() #Acomoda años en ascendente y goles en descendente y regresa solo el primer equipo de cada año

most_goals_per_year_team
```
